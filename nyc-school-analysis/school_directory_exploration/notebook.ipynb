**Loading Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Cleaning the columns names**

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace(r'[^\w_]', '', regex=True)
df.columns
```

**Visualizations**

```python
# Count of unique 'dbn' per 'borough'
dbn_count_per_borough = df.groupby('borough')['dbn'].nunique()

# Plotting the bar chart
dbn_count_per_borough.plot(kind='bar', color='skyblue', figsize=(8, 6))
plt.title('Number of Unique Schools (DBN) per Borough')
plt.xlabel('Borough')
plt.ylabel('Count of Unique DBNs')
plt.xticks(rotation=45)
plt.show()
```

- Brooklyn has the highest number of unique schools (DBNs), followed closely by the Bronx and Manhattan.
- Staten Island has significantly fewer schools compared to the other boroughs.

**ðŸ•˜ 6. School Day Length Distribution**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert time strings to datetime.time objects
df['start_time_dt'] = pd.to_datetime(df['start_time'], format='%I:%M %p', errors='coerce')
df['end_time_dt'] = pd.to_datetime(df['end_time'], format='%I:%M %p', errors='coerce')

# Calculate duration in hours
df['school_day_hours'] = (df['end_time_dt'] - df['start_time_dt']).dt.total_seconds() / 3600

# Drop invalid entries (if start or end time was missing or invalid)
df = df[df['school_day_hours'].notna()]

# Plot
sns.histplot(df['school_day_hours'], bins=10, kde=True)
plt.title('Distribution of School Day Length (in Hours)')
plt.xlabel('Hours')
plt.ylabel('Number of Schools')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud
import folium

# Load your cleaned dataset here
# df = pd.read_csv("cleaned_high_school_directory.csv")  # Replace with your actual file

# ========== ðŸ« 4. Enrichment & Partnerships ==========

# Count non-null values across partner_* columns
partner_cols = [
    'partner_cbo', 'partner_hospital', 'partner_highered', 'partner_cultural',
    'partner_nonprofit', 'partner_corporate', 'partner_financial', 'partner_other'
]

df['num_partnerships'] = df[partner_cols].notna().sum(axis=1)

# Top 10 schools by partnerships
top_partners = df.nlargest(10, 'num_partnerships')

plt.figure(figsize=(10, 6))
sns.barplot(data=top_partners, x='num_partnerships', y='school_name', palette='viridis')
plt.title('Top 10 Schools by Number of Partnerships')
plt.xlabel('Partnership Count')
plt.ylabel('School Name')
plt.tight_layout()
plt.show()

# Word cloud of extracurricular activities
activity_text = " ".join(df['extracurricular_activities'].dropna().astype(str))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(activity_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Extracurricular Activities (Word Cloud)")
plt.show()

# ========== ðŸ€ 5. Sports Opportunities ==========

# Count sports per gender
df['num_boys_sports'] = df['psal_sports_boys'].fillna('').apply(lambda x: len(x.split(',')) if x else 0)
df['num_girls_sports'] = df['psal_sports_girls'].fillna('').apply(lambda x: len(x.split(',')) if x else 0)
df['num_coed_sports'] = df['psal_sports_coed'].fillna('').apply(lambda x: len(x.split(',')) if x else 0)

# Bar plot
sports_sum = {
    "Boys": df['num_boys_sports'].sum(),
    "Girls": df['num_girls_sports'].sum(),
    "Co-ed": df['num_coed_sports'].sum()
}
plt.figure(figsize=(8, 5))
sns.barplot(x=list(sports_sum.keys()), y=list(sports_sum.values()), palette='coolwarm')
plt.title("Total Sports by Gender")
plt.ylabel("Number of Sports")
plt.show()

# PSAL vs Non-PSAL vs Both
df['has_psal'] = df[['psal_sports_boys', 'psal_sports_girls', 'psal_sports_coed']].notna().any(axis=1)
df['has_school_sport'] = df['school_sports'].notna()

psal_only = df[(df['has_psal']) & (~df['has_school_sport'])].shape[0]
non_psal_only = df[(~df['has_psal']) & (df['has_school_sport'])].shape[0]
both_sports = df[(df['has_psal']) & (df['has_school_sport'])].shape[0]

plt.figure(figsize=(6, 6))
plt.pie([psal_only, non_psal_only, both_sports],
        labels=["PSAL Only", "Non-PSAL Only", "Both"],
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'width': 0.4})
plt.title("Comparison of PSAL vs Non-PSAL Sports")
plt.show()
```
